```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

'Country code' is a common column in both the datasets.

```python
final_df = pd.merge(df,df_country,on='Country Code', how='right')
final_df
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
 final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart
plt.pie(country_val)
plt.legend(country_names)
```

```python
#top 3 companies that use zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

## 2. Numeric Values

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).first()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation

1.When rating is between 4.5 to 4.9---> Excellent
2.When rating is between 4.0 to 3.4---> very good
3.When rating is between 3.5 to 3.4---> good
4.When rating is between 3.0 to 3.4---> average
5.When rating is between 2.5 to 2.9---> average
6.When rating is between 2.0 to 2.4---> power

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] =(12,6)#increases the size
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',palette=['white','red','orange','yellow', 'green'],data=ratings)
```

```python
#COUNT PLOT
sns.countplot(x='Rating color',palette=['white','red','orange','yellow', 'green'],data=ratings)
```
